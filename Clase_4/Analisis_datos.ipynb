```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = list(range(1,6))
y = list(range(1,14, 3))
```

```python
x
y
```

```python
plt.plot(x,y)
plt.ylabel('Eje y')
plt.xlabel('Eje x')
plt.title('Grafico lineas')
plt.show()
```

```python
plt.scatter(x,y)
plt.ylabel('Eje y')
plt.xlabel('Eje x')
plt.title('Grafico Dispersión')
plt.show()
```

```python
import numpy as np 
datos = np.random.rand(1000)

plt.hist(datos, bins= 10, color='skyblue', edgecolor = 'black')
plt.show()
```
