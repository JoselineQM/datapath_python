```python
data ={
    'A':[1,2,3,4],
    'B':[0,4,5,7],
    'C':[3,5,7,8],
}

df1 = pd.DataFrame(data)
```

```python
data ={
    'C':[1,2,3,4,5],
    'D':[3, 4, 6, 8,2],
    'E':[1,2,3,5,2]
}

df2 = pd.DataFrame(data)
```

```python
pd.concat((df1, df2), axis=0)
```

```python
pd.concat((df1, df2), axis=1)
```

```python
pd.merge(df1, df2, on='C')
```

```python
df1.merge(df2, on='C')
```

```python
df1.merge(df2, on='C', how='left')#left join
```

```python
df1.merge(df2, on='C', how='right')#right join
```

```python
df1.merge(df2, on='C', how='right')#right join
```

```python
df1.merge(df2, left_on='C', right_on='D')#inner join 
```
